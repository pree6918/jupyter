```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.visualization import astropy_mpl_style
plt.style.use(astropy_mpl_style)
from astropy.utils.data import get_pkg_data_filename
from astropy.io import fits
```

```python
ranom1=321.485826-360 #Taken from PDS
decnom1=-16.447116
ranom2=321.486113-360
decnom2=-16.447078
ranom3=321.486400-360
decnom3=-16.447040
ranom4=321.486688-360
decnom4=-16.447003
offra=0.00011
offdec=0.00004
angint=.00006
angback=.00012

image_file = get_pkg_data_filename(
    "stage3/d1_fringe_nav/Level3_ch2-medium_s3d_nav.fits")
#image_file = "stage3/d1_fringe_nav/Level3_ch1-short_s3d_nav.fits"
hdul=fits.open(image_file)
hdr=hdul[1].header
qub1s1=fits.getdata(image_file, ext=1).astype(float)
npix1=hdr['NAXIS1']
npix2=hdr['NAXIS2']
nwave=hdr['NAXIS3']
wave0=hdr['CRVAL3']
pixsr=hdr['PIXAR_SR']
dra=hdr['CDELT1']
ddec=hdr['CDELT2']
dwave=hdr['CDELT3']
RA0=hdr['CRVAL1']
DEC0=hdr['CRVAL2']
PIX10=hdr['CRPIX1']
PIX20=hdr['CRPIX2']
PC11=hdr['PC1_1']
PC12=hdr['PC1_2']
PC21=hdr['PC2_1']
PC22=hdr['PC2_2']
ras1=np.zeros([npix2,npix1])
decs1=np.zeros([npix2,npix1])
for i in range(npix1):
    ras1[:,i]=RA0+dra*PC11*(i-PIX10)+dra*PC12*(np.linspace(1,npix2,npix2)-PIX20)-offra
    decs1[:,i]=DEC0+ddec*PC21*(i-PIX10)+ddec*PC22*(np.linspace(1,npix2,npix2)-PIX20)-offdec
wave1s=np.linspace(wave0,wave0+nwave*dwave,nwave)
plt.plot(ras1-ranom1, decs1-decnom1, 'bo')
#print(hdr)
image_file = get_pkg_data_filename(
    "stage3/d2_fringe_nav/Level3_ch2-medium_s3d_nav.fits")
hdul=fits.open(image_file)
hdr=hdul[1].header
qub1s2=fits.getdata(image_file, ext=1).astype(float)
npix1=hdr['NAXIS1']
npix2=hdr['NAXIS2']
nwave=hdr['NAXIS3']
wave0=hdr['CRVAL3']
dra=hdr['CDELT1']
ddec=hdr['CDELT2']
dwave=hdr['CDELT3']
RA0=hdr['CRVAL1']
DEC0=hdr['CRVAL2']
PIX10=hdr['CRPIX1']
PIX20=hdr['CRPIX2']
PC11=hdr['PC1_1']
PC12=hdr['PC1_2']
PC21=hdr['PC2_1']
PC22=hdr['PC2_2']
ras2=np.zeros([npix2,npix1])
decs2=np.zeros([npix2,npix1])
for i in range(npix1):
    ras2[:,i]=RA0+dra*PC11*(i-PIX10)+dra*PC12*(np.linspace(1,npix2,npix2)-PIX20)-offra
    decs2[:,i]=DEC0+ddec*PC21*(i-PIX10)+ddec*PC22*(np.linspace(1,npix2,npix2)-PIX20)-offdec
wave1s=np.linspace(wave0,wave0+nwave*dwave,nwave)
plt.plot(ras2-ranom2, decs2-decnom2, 'go')
#print(hdr)
image_file = get_pkg_data_filename(
    "stage3/d3_fringe_nav/Level3_ch2-medium_s3d_nav.fits")
hdul=fits.open(image_file)
hdr=hdul[1].header
qub1s3=fits.getdata(image_file, ext=1).astype(float)
npix1=hdr['NAXIS1']
npix2=hdr['NAXIS2']
nwave=hdr['NAXIS3']
wave0=hdr['CRVAL3']
dra=hdr['CDELT1']
ddec=hdr['CDELT2']
dwave=hdr['CDELT3']
RA0=hdr['CRVAL1']
DEC0=hdr['CRVAL2']
PIX10=hdr['CRPIX1']
PIX20=hdr['CRPIX2']
PC11=hdr['PC1_1']
PC12=hdr['PC1_2']
PC21=hdr['PC2_1']
PC22=hdr['PC2_2']
ras3=np.zeros([npix2,npix1])
decs3=np.zeros([npix2,npix1])
for i in range(npix1):
    ras3[:,i]=RA0+dra*PC11*(i-PIX10)+dra*PC12*(np.linspace(1,npix2,npix2)-PIX20)-offra
    decs3[:,i]=DEC0+ddec*PC21*(i-PIX10)+ddec*PC22*(np.linspace(1,npix2,npix2)-PIX20)-offdec
wave1s=np.linspace(wave0,wave0+nwave*dwave,nwave)
plt.plot(ras3-ranom3, decs3-decnom3, 'yo')
#print(hdr)
image_file = get_pkg_data_filename(
    "stage3/d4_fringe_nav/Level3_ch2-medium_s3d_nav.fits")
hdul=fits.open(image_file)
hdr=hdul[1].header
qub1s4=fits.getdata(image_file, ext=1).astype(float)
npix1=hdr['NAXIS1']
npix2=hdr['NAXIS2']
nwave=hdr['NAXIS3']
wave0=hdr['CRVAL3']
dra=hdr['CDELT1']
ddec=hdr['CDELT2']
dwave=hdr['CDELT3']
RA0=hdr['CRVAL1']
DEC0=hdr['CRVAL2']
PIX10=hdr['CRPIX1']
PIX20=hdr['CRPIX2']
PC11=hdr['PC1_1']
PC12=hdr['PC1_2']
PC21=hdr['PC2_1']
PC22=hdr['PC2_2']
ras4=np.zeros([npix2,npix1])
decs4=np.zeros([npix2,npix1])
for i in range(npix1):
    ras4[:,i]=RA0+dra*PC11*(i-PIX10)+dra*PC12*(np.linspace(1,npix2,npix2)-PIX20)-offra
    decs4[:,i]=DEC0+ddec*PC21*(i-PIX10)+ddec*PC22*(np.linspace(1,npix2,npix2)-PIX20)-offdec
wave1s=np.linspace(wave0,wave0+nwave*dwave,nwave)
plt.plot(ras4-ranom4, decs4-decnom4, 'ro')
plt.xlim([.0011,-.0011])
plt.ylim([-.0011,.0011])
#print(hdr)
plt.show()
```

```python
dist1=np.sqrt((ras1-ranom1)**2+(decs1-decnom1)**2)
dist2=np.sqrt((ras2-ranom2)**2+(decs2-decnom2)**2)
dist3=np.sqrt((ras3-ranom3)**2+(decs3-decnom3)**2)
dist4=np.sqrt((ras4-ranom4)**2+(decs4-decnom4)**2)
range1=dist1-dist2
range1[(dist1<angint)]=1
range1[(dist1>angint) & (dist1<angback)]=0.5
range2=dist1-dist2
range2[(dist2<angint)]=1
range2[(dist2>angint) & (dist2<angback)]=0.5
range3=dist1-dist2
range3[(dist3<angint)]=1
range3[(dist3>angint) & (dist3<angback)]=0.5
range4=dist1-dist2
range4[(dist4<angint)]=1
range4[(dist4>angint) & (dist4<angback)]=0.5
plt.imshow(range1)
plt.show()
plt.imshow(np.nanmean(qub1s1-qub1s2, axis=0)-np.nanmedian(np.nanmean(qub1s1-qub1s2, axis=0)), vmin=-200, vmax=200)
plt.show()
plt.imshow(range2)
plt.show()
plt.imshow(np.nanmean(qub1s2-qub1s1, axis=0)-np.nanmedian(np.nanmean(qub1s2-qub1s1, axis=0)), vmin=-200, vmax=200)
plt.show()

plt.imshow(range3)
plt.show()
plt.imshow(np.nanmean(qub1s3-qub1s2, axis=0)-np.nanmedian(np.nanmean(qub1s3-qub1s2, axis=0)), vmin=-200, vmax=200)
plt.show()
plt.imshow(range4)
plt.show()
plt.imshow(np.nanmean(qub1s4-qub1s1, axis=0)-np.nanmedian(np.nanmean(qub1s4-qub1s1, axis=0)), vmin=-200, vmax=200)
plt.show()
```

```python
specs=np.zeros([4,nwave])
foo1=(dist1<angint)
foob1=(dist1>angint) & (dist1<angback)
foo2=(dist2<angint)
foob2=(dist2>angint) & (dist2<angback)
foo3=(dist3<angint)
foob3=(dist3>angint) & (dist3<angback)
foo4=(dist4<angint)
foob4=(dist4>angint) & (dist4<angback)
for i in range(nwave):
    dat1=qub1s1[i,:,:]-qub1s2[i,:,:]
    specs[0,i]=(np.nanmean(dat1[foo1])-np.nanmean(dat1[foob1]))*sum(sum(foo1))*pixsr*1e12
    dat2=qub1s2[i,:,:]-qub1s1[i,:,:]
    specs[1,i]=(np.nanmean(dat2[foo2])-np.nanmean(dat2[foob2]))*sum(sum(foo2))*pixsr*1e12
    dat3=qub1s3[i,:,:]-qub1s4[i,:,:]
    specs[2,i]=(np.nanmean(dat3[foo3])-np.nanmean(dat3[foob3]))*sum(sum(foo3))*pixsr*1e12
    dat4=qub1s4[i,:,:]-qub1s3[i,:,:]
    specs[3,i]=(np.nanmean(dat4[foo4])-np.nanmean(dat4[foob4]))*sum(sum(foo4))*pixsr*1e12
plt.plot(wave1s,specs[0,:], 'red')
plt.plot(wave1s,specs[1,:], 'blue')
plt.plot(wave1s,specs[2,:], 'green')
plt.plot(wave1s,specs[3,:], 'yellow')
plt.plot(wave1s,np.nanmean(specs[:,:], axis=0), 'k')
plt.ylabel('Flux ($\mu$Jy)')
plt.xlabel('Wavelenegth (microns)')
plt.show()
```
